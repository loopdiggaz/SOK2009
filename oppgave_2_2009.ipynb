Vi har en variabel X som har et utfallsrom x âˆˆ {1, 2, 3} der utfall har sannsynlighet p(X = x) = x/6 for alle utfall i utfallsrommet.

1. Kan du tegne opp fordelingen til f(x) for verdiene i utfallsrommet ? Hva kaller vi denne fordelingen?

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

probability = [1/6,2/6,3/6]

x = [1,2,3]

fig, ax = plt.subplots()
ax.bar(x,probability)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_title('Fordeling f(x)');
```

Dette er en diskret sannsynlighetsfordeling. Varibalene tar et endelig sett med verdier, hvor hver av verdiene har en klart definert sannsynlighet 

2. Kan du tegne den kumulative sannsynlighetsfordelingen F(X) for X og $ X^3 $?

```python
kum_probability = [1/6,1/6+1/3,1/6+1/3+1/2]

fig, ax = plt.subplots()
ax.bar(x,kum_probability)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_title('Kumulativ fordeling for f(x)');
```

```python
x_3 = [1,8,27]

fig, ax = plt.subplots()
ax.bar(x_3,kum_probability)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_title('Kumulativ fordeling for $ f(x)^3 $');
```
