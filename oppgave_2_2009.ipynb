2. Kan du tegne den kumulative sannsynlighetsfordelingen $ F(X) $ for $ X $ og $ X^3 $?

```python
import matplotlib.pyplot as plt
import numpy as np

# Definerer verdier for X
x_values = np.array([1, 2, 3])

# Sannsynligheter for X og X^3
p_x = x_values / 6
p_x3 = (x_values**3) / np.sum(x_values**3)

# Kumulative fordelinger
cdf_x = np.cumsum(p_x)
cdf_x3 = np.cumsum(p_x3)

# Oppretter figur og akser
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plott for X
ax[0].step(x_values, cdf_x, where='mid', linestyle='--', marker='o')
ax[0].set_title('Kumulativ Sannsynlighetsfordeling for X')
ax[0].set_xlabel('X')
ax[0].set_ylabel('F(X)')
ax[0].grid(True)
ax[0].set_ylim(0, 1.1)
ax[0].set_xticks(x_values)

# Plott for X^3
ax[1].step(x_values, cdf_x3, where='mid', linestyle='-', marker='o', color='r')
ax[1].set_title('Kumulativ Sannsynlighetsfordeling for X^3')
ax[1].set_xlabel('X')
ax[1].set_ylabel('F(X^3)')
ax[1].grid(True)
ax[1].set_ylim(0, 1.1)
ax[1].set_xticks(x_values)

plt.tight_layout()
plt.show()
```

4. Kan du regne ut korrelasjon, kovarians, varians og standardavik for $ (Y_1, Y_2) der  Y_1 = 3 + 5X og Y_2 = 4 + 6X^3 $ og
$ (X, X^3) $ for verdiene x âˆˆ {1, 2, 3} . Kan du forklare svaret du fÃ¥r med egne ord?

```python
def y(x):
    return 3 + 5 * x

def y2(x):
    return 4 + 6 * x**3

# Definerer x-verdiene
x_verdier = [1, 2, 3]

# Bruker funksjonene pÃ¥ x-verdiene
y_verdier = [y(x) for x in x_verdier]
y2_verdier = [y2(x) for x in x_verdier]

import pandas as pd

# Lager et datasett
df = pd.DataFrame({'y': y_verdier, 'y2': y2_verdier})

df.corr(method= 'pearson')
korrelasjon = 0.966

print(f'Korrelasjon mellom y og y2 er:')
display(korrelasjon)

kovariansmatrise = df.cov()

kovariansmatrise
```
