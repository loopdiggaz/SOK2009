```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dice(n):
    rolls = []
    for i in range(n):
        two_dice = ( np.random.randint(1, 7) + np.random.randint(1, 7) )
        rolls.append(two_dice)
    return rolls

data = dice(100000)
# print(data)
bins = np.arange(14) - 0.5 
plt.hist(data, bins, color="blue", rwidth=0.85)
plt.grid(True)
plt.title("2 dice rolling histogram")
plt.xlabel("Sum of 2 dice")
plt.ylabel("frequency of rolling the number")
plt.xticks(range(1,14))
# Set axes limit
plt.xlim(1,13)
```
